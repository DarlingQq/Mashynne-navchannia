```python
import numpy as np
```

```python
# Завдання 1
# Створіть одновимірний масив (вектор) з першими 20-ма натуральними числами та виведіть його значення.
z1 = np.arange(1, 21)
print("Завдання 1:\n", z1, "\n")
```

```python
# Завдання 2
# Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його випадковими натуральними числами та виведіть його значення.
z2 = np.random.randint(1, 100, (3, 3))
print("Завдання 2:\n", z2, "\n")
```

```python
# Завдання 3
# Створіть масив розміром 5x5, заповніть його випадковими дійсними числами в діапазоні від -1 до 1 та виведіть його значення.
z3 = np.random.uniform(-1, 1, (5, 5))
print("Завдання 3:\n", z3, "\n")
```

```python
# Завдання 4
# Створіть масив розміром 5x5, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть середнє значення елементів.
z4 = np.random.rand(5, 5)
print("Завдання 4 (матриця):\n", z4)
print("Середнє значення:", np.mean(z4), "\n")
```

```python
# Завдання 5
# Створіть два одновимірних масиви розміром 5, заповніть їх випадковими цілими числами від -10 до 10 та виконайте на них поелементні операції додавання, віднімання та множення.
a5 = np.random.randint(-10, 11, 5)
b5 = np.random.randint(-10, 11, 5)
print("Завдання 5:")
print("a:", a5)
print("b:", b5)
print("Додавання:", a5 + b5)
print("Віднімання:", a5 - b5)
print("Множення:", a5 * b5, "\n")
```

```python
# Завдання 6
# Створіть дві матриці 2x2, заповніть їх довільними дійсними числами від 3 до 10 та знайдіть їх добуток.
A6 = np.random.uniform(3, 10, (2, 2))
B6 = np.random.uniform(3, 10, (2, 2))
C6 = np.dot(A6, B6)
print("Завдання 6:")
print("A:\n", A6)
print("B:\n", B6)
print("A*B:\n", C6, "\n")
```

```python
# Завдання 7
# Створіть дві матриці 2x3 та 3x2, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та перемножте їх між собою.
A7 = np.random.randint(1, 6, (2, 3))
B7 = np.random.randint(1, 6, (3, 2))
C7 = np.dot(A7, B7)
print("Завдання 7:")
print("A:\n", A7)
print("B:\n", B7)
print("A*B:\n", C7, "\n")
```

```python
# Завдання 8
# Створіть матрицю 3x3, заповніть її випадковими цілими числами від 1 до 12 та знайдіть її обернену матрицю.
A8 = np.random.randint(1, 13, (3, 3))
invA8 = np.linalg.inv(A8)
print("Завдання 8:\nA:\n", A8)
print("A^(-1):\n", invA8, "\n")
```

```python
# Завдання 9
# Створіть матрицю 4x4, заповніть її випадковими дійсними числами в діапазоні від -8 до 8 та транспонуйте її.
A9 = np.random.uniform(-8, 8, (4, 4))
print("Завдання 9:\nA:\n", A9)
print("A^T:\n", A9.T, "\n")
```

```python
# Завдання 10
# Створіть матрицю 3x4 та вектор 4x1, заповніть їх випадковими цілими числами від 1 до 10 та знайдіть середнє значення елементів матриці.
A10 = np.random.randint(1, 11, (3, 4))
b10 = np.random.randint(1, 11, (4, 1))
print("Завдання 10:")
print("A:\n", A10)
print("b:\n", b10)
print("Середнє значення елементів матриці A:", np.mean(A10), "\n")
```

```python
# Завдання 11
# Створіть матрицю розміром 2x3 та вектор розміром 3x1, заповніть матрицю випадковими цілими числами в діапазоні від 1 до 10, а вектор - випадковими дійсними числами від 0 до 1, і знайдіть добуток матриці на вектор.
A11 = np.random.randint(1, 11, (2, 3))
b11 = np.random.rand(3, 1)
prod11 = A11.dot(b11)
print("Завдання 11 — матриця A (2x3):")
print(A11)
print("Вектор b (3x1):")
print(b11)
print("Добуток A * b (2x1):")
print(prod11)
print()
```

```python
# Завдання 12
# Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть їхню різницю.
A12 = np.random.randint(1, 11, (3, 3))
B12 = np.random.randint(1, 11, (3, 3))
diff12 = A12 - B12
print("Завдання 12 — матриці A і B:")
print(A12)
print(B12)
print("A − B:")
print(diff12)
print()
```

```python
# Завдання 13
# Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від -5 до 5 та знайдіть їх добуток.
A13 = np.random.randint(-5, 6, (3, 3))
B13 = np.random.randint(-5, 6, (3, 3))
prod13 = np.dot(A13, B13)
print("Завдання 13 — матриці A і B:")
print(A13)
print(B13)
print("A * B:")
print(prod13)
print()
```

```python
# Завдання 14
# Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від -5 до 5 та знайдіть суму елементів матриці.
A14 = np.random.randint(-5, 6, (3, 3))
sum14 = np.sum(A14)
print("Завдання 14 — матриця A:")
print(A14)
print("Сума елементів матриці A:", sum14)
print()
```

```python
# Завдання 15
# Створіть матрицю розміром 4x4, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть вектор-стовпчик,
# що містить добуток елементів кожного рядка матриці.
A15 = np.random.rand(4, 4)
prod_rows_15 = np.prod(A15, axis=1).reshape(4, 1)
print("Завдання 15 — матриця A (4x4):")
print(A15)
print("Вектор-стовпчик (добуток рядків):")
print(prod_rows_15)
```
