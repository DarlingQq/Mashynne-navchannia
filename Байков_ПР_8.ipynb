```python
features = df[['hdlngth', 'skullw', 'totlngth', 'taill', 'footlgth', 'earconch', 'eye', 'chest', 'belly']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Припустимо, що у тебе є датафрейм df з числовими ознаками
X = df.values  # або df[['col1', 'col2', ...]].values для вибраних колонок

silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Візуалізація результатів
plt.plot(k_range, silhouette_scores, marker='o', color='darkgreen')
plt.title("Метод силуетів для визначення оптимального k")
plt.xlabel("Кількість кластерів (k)")
plt.ylabel("Середній силует-скір")
plt.grid(True)
plt.show()
```

```python
# Метод силуетів для визначення оптимальної кількості кластерів
silhouette_scores = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    score = silhouette_score(X_scaled, kmeans.labels_)
    silhouette_scores.append(score)

# Побудова графіка силуетів
plt.figure(figsize=(10, 7))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--', color='crimson', linewidth=2, markersize=8)
plt.title('Метод силуетів для визначення кількості кластерів', fontsize=14)
plt.xlabel('Кількість кластерів', fontsize=12)
plt.ylabel('Силуетний коефіцієнт', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Використання KMeans для кластеризації (3 кластери)
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Візуалізація кластерів
plt.figure(figsize=(12, 8))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=clusters, palette='viridis', legend='full')

# Додавання центроїдів
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Центроїди', marker='X')

# Настроювання графіка
plt.title('Кластеризація KMeans з 3 кластерами')
plt.xlabel('Ознака 1 (масштабована)')
plt.ylabel('Ознака 2 (масштабована)')
plt.legend(title='Кластер')
plt.grid(True, alpha=0.3)
plt.show()
```
